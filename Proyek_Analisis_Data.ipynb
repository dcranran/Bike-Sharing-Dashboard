## Analisis Lanjutan (Clustering)

Penyewaan sepeda akan dibagi menjadi 3 kelompok
1. Rendah (Low Demand)
2. Sedang (Medium Demand)
3. Tinggi (High Demand)


```python
bins = [0, 2000, 4000, day_df['cnt'].max()]
labels = ["Low Demand", "Medium Demand", "High Demand"]
day_df["demand_category"] = pd.cut(day_df["cnt"], bins=bins, labels=labels, include_lowest=True)
demand_distribution = day_df["demand_category"].value_counts()
demand_distribution
```

```python
plt.pie(demand_distribution, labels=demand_distribution.index, autopct='%1.1f%%', colors=["lightblue", "lightyellow", "pink"])
plt.title("Distribusi Clustering Penyewaan Sepeda")
```

```python
plt.figure(figsize=(8, 6))
color = ["#D3D3D3", "#D3D3D3", "#EB5679"]

sns.barplot(x=demand_distribution.index, y=demand_distribution.values, hue=demand_distribution.index, palette=color)
plt.title("Distribusi Clustering Penyewaan Sepeda (Manual Binning)")
plt.xlabel("Kategori Permintaan")
plt.ylabel("Jumlah Hari")
plt.show()
```

## Conclusion

- **Conclusion pertanyaan 1** 

  Berdasarkan analisis yang telah dilakukan, dapat disimpulkan bahwa cuaca dan musim memiliki pengaruh signifikan terhadap permintaan penyewaan sepeda. Cuaca cerah meningkatkan jumlah penyewaan, sedangkan kondisi cuaca yang lebih buruk seperti mendung, hujan ringan, dan terutama hujan lebat menyebabkan penurunan jumlah penyewa secara drastis. Suhu memiliki hubungan positif dengan jumlah penyewaan, di mana semakin tinggi suhu, semakin banyak pengguna yang menyewa sepeda. Namun, suhu ekstrem yang terlalu panas atau terlalu dingin dapat menurunkan kenyamanan pengguna. Kelembaban dan kecepatan angin berhubungan negatif dengan jumlah penyewaan, di mana kelembaban tinggi dan angin kencang membuat pengguna cenderung mengurangi aktivitas bersepeda. Musim gugur memiliki rata-rata penyewaan tertinggi, diikuti oleh musim panas, sementara musim semi memiliki rata-rata penyewaan terendah. Hal ini menunjukkan bahwa pengguna lebih aktif bersepeda saat kondisi cuaca lebih nyaman dan stabil.

  Dari hasil tersebut, dapat disusun beberapa rekomendasi berikut:
  - Meningkatkan jumlah sepeda saat musim gugur dan musim panas untuk memenuhi permintaan yang lebih tinggi.
  - Mengoptimalkan strategi harga dan promosi saat cuaca buruk atau musim dingin, misalnya dengan diskon khusus atau layanan tambahan seperti jas hujan bagi pengguna.
  - Memantau prakiraan cuaca untuk mengantisipasi permintaan, sehingga operasional dapat disesuaikan untuk efisiensi yang lebih baik.

- **Conclusion pertanyaan 2** 

  Berdasarkan analisis yang telah dilakukan, dapat disimpulkan bahwa permintaan penyewaan sepeda memiliki pola musiman, harian, dan jam tertentu yang memengaruhi jumlah penyewa
